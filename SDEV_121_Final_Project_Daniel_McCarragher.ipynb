##Is the number of smokers in the United States increasing or decreasing since 1980?

As shown from the scatterplot shown below, it is quite obvious that the number of smokers, men and women included, has been consistently decreasing since 1980.The trend in the decrease seems to be consistent year after year.

```python
import seaborn as sns
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt

## This code was provided from SDEV121 Final Project Review Notebook
def with_commas(x, not_used):
    '''
    This function takes a numerical input x and returns it as a string with no decimal places
    and with commas as thousands separators
    '''
    return f"{x:,.0f}"

#create line graph
ax = sns.lineplot(data=United_States_Smoking_df, x='Year', y='Smokers Total', hue='Country')

# Set y label
ax.set(ylabel="Smokers Total (Millions)")

# Set x-axis tick labels
ax.set_xticks(United_States_Smoking_df['Year'])

#rotate x ticks for better readability
plt.xticks(rotation=90)


ax.yaxis.set_major_formatter(FuncFormatter(with_commas))

```

##Import JSON file into pandas dataframe

```python
cars_df = pd.read_json("cars.json")
#show the unedited json file in the dataframe
cars_df
```

##Store flattened JSON data in dataframe.

```python
import json

#open the json file
with open("cars.json", "r") as file:
    data = json.load(file)

#flatten the JSON data
flattened_cars_df = pd.json_normalize(data, sep=' ', max_level = 2)
flattened_cars_df
```

```python
#Rename Columns making them easier to read

#I used both renaming techniques in this project
flattened_cars_df.columns = [
    'Height', 'Length', 'Width', 'Driveline', 'Engine Type', 'Hybrid',
    'Number of Forward Gears', 'Transmission', 'Horsepower', 'Torque',
    'City mpg', 'Fuel Type', 'Highway mpg', 'Classification', 'ID',
    'Make', 'Model Year', 'Year'
]
```

Show the columns new names after being flattened.


```python
flattened_cars_df.columns
```

##Create barchart with mean of each brand and city MPG




Shows the mean of City MPG for each individual Car Manufacturer. Majority of data hovers between 15 and 20 mpg, with a few outliers around 25. The box plot below shows that the mean for all data points is approxiametly 17.5 mpg.

```python
#find the mean of each vehichle City MPG dependent on each make of vehicle
mean_city_mpg = flattened_cars_df.groupby("Make")["City mpg"].mean().reset_index()


ax = mean_city_mpg.plot(x='Make', y='City mpg', kind='bar', figsize=(12, 6))
ax.set(title="City MPG per Car Brand", xlabel='Brands')
```

```python
#create boxplot
ax = mean_city_mpg.plot(x='Make', y='City mpg', kind='box', figsize=(12, 6))
#title and labels
ax.set(title="City MPG per Car Brand", xlabel='Brands')
```

##Create scatterplot that shows the correlation between an increase in horsepower, and a decrease in Miles Per Gallon

With using the scatterplot as a visual, it becomes prevalent that within the dataset, as the Horsepower of the vehichle increase, the MPG significantly increases. This can be likely drawn from the conclusion that with an increase in power from the engine, this requires more fuel.

```python
# Select the columns
engine_city_mpg_data = flattened_cars_df[['Make',
                                          'Horsepower',
                                          'Torque',
                                          'City mpg']]

#create scatterplot
ax = engine_city_mpg_data.plot(x='Horsepower',
                               y='City mpg', kind='scatter', figsize=(10, 6))

#labels and title
ax.set(xlabel = "Horsepower", ylabel="City MPG", title='Horsepower vs City MPG')


```
